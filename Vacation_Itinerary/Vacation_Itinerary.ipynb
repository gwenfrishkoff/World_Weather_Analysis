## Import Source Data

```python
# Read the WeatherPy_vacation.csv into a DataFrame.
vacation_df = pd.read_csv("WeatherPy_vacation.csv")
vacation_df.head()
```

## Retrieve Data for Marker Layer Maps

```python
# Using the template add city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Hotel Name</dt><dd>{Hotel Name} at {Max Temp}</dd>
"""

# Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in vacation_df.iterrows()]

# Get the latitude and longitude from each row and store in a new DataFrame.
locations = vacation_df[["Lat", "Lng"]]
```

## Create the Marker Layer Map

```python
# Add a marker layer for each city to the map.
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

# Display the figure
fig = gmaps.figure()
fig.add_layer(marker_layer)

fig
```

## Retrieve data for direction layer map

```python
# Restrict DF to US destinations
vacation_df.loc[vacation_df["Country"] == "US"]
```

```python
# From the map above pick 4 cities and create a vacation itinerary route to travel between the four cities. 

# Create DataFrames for each city by filtering the 'vacation_df' using the loc method. 
# Hint: The starting and ending city should be the same city.

vacation_start = vacation_df.loc[vacation_df["City"] == "Eureka"]
vacation_end = vacation_df.loc[vacation_df["City"] == "Eureka"]
vacation_stop1 = vacation_df.loc[vacation_df["City"] == "San Jose"]
vacation_stop2 = vacation_df.loc[vacation_df["City"] == "Carlsbad"] 
vacation_stop3 = vacation_df.loc[vacation_df["City"] == "Morro Bay"] 
```

```python
# Get the latitude-longitude pairs as tuples from each city DataFrame using the to_numpy function.
#start = vacation_start[["Lat", "Lng"]].to_numpy()
#end = vacation_end[["Lat", "Lng"]].to_numpy()
#stop1 = vacation_stop1[["Lat", "Lng"]].to_numpy()
#stop2 = vacation_stop2[["Lat", "Lng"]].to_numpy()
#stop3 = vacation_stop3[["Lat", "Lng"]].to_numpy()

# NOTE: Alt syntax using list indexing (yields lists, not tuples... not sure how to combine to_numpy w/indexing)
start = vacation_start["Lat"].values[0],vacation_start["Lng"].values[0]
end = vacation_end["Lat"].values[0],vacation_end["Lng"].values[0]
stop1 = vacation_stop1["Lat"].values[0],vacation_stop1["Lng"].values[0]
stop2 = vacation_stop2["Lat"].values[0],vacation_stop2["Lng"].values[0]
stop3 = vacation_stop3["Lat"].values[0],vacation_stop3["Lng"].values[0]

print(start)
print(end)
print(stop1)
print(stop2)
print(stop3)
```

## Create a direction layer map

```python
# Create a direction layer map using the start and end latitude-longitude pairs,
# and stop1, stop2, and stop3 as the waypoints. The travel_mode should be "DRIVING", "BICYCLING", or "WALKING".
fig = gmaps.figure()

vacation_itinerary = gmaps.directions_layer(
    start,end,waypoints=[stop1,stop2,stop3],
    travel_mode="DRIVING"
)

fig.add_layer(vacation_itinerary)
fig
```

## Create the WeatherPy_travel (marker layer) map

```python
#  Combine the four city DataFrames into one DataFrame using the concat() function.
itinerary_df = pd.concat([vacation_start,vacation_stop1,vacation_stop2,vacation_stop3],ignore_index=True)
itinerary_df
```

```python
# Using the template add city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Hotel Name</dt><dd>{Hotel Name} at {Max Temp}</dd>
"""

# Get the data from each row and add it to the formatting template and store the data in a list.
travel_info = [info_box_template.format(**row) for index, row in itinerary_df.iterrows()]

# Get the latitude and longitude from each row and store in a new DataFrame.
locations = itinerary_df[["Lat", "Lng"]]
```

```python
# Add a marker layer for each city to the map. 
marker_layer = gmaps.marker_layer(locations,info_box_content=travel_info)

# Define fig object
fig = gmaps.figure()

# Add the layer to the map
fig.add_layer(marker_layer)

# Display the figure
fig
```
